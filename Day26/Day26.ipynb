```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Đọc và hiển thị ảnh
img = cv2.imread('dataset/1.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img , cmap="gray")
plt.show()
```

```python
#Đọc và hiển thị ảnh
img = cv2.imread('dataset/1.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Tăng độ sáng
new_img = img.astype(np.float32)+50
new_img = np.clip(new_img , 0 , 255)
new_img = new_img.astype(np.uint8)

# Hiển thị hình ảnh gốc
plt.subplot(2, 2, 1) # Lưới 2x2, vị trí 1
plt.imshow(img)
plt.title('Origin Image')

# Hiển thị hình ảnh tăng độ sáng
plt.subplot(2, 2, 2) # Lưới 2x2, vị trí 2
plt.imshow(new_img)
plt.title('Brightness Image')

new_img = img.astype(np.float32)-80
new_img = np.clip(new_img , 0 , 255)
new_img = new_img.astype(np.uint8)

# Hiển thị hình ảnh giảm độ sáng
plt.subplot(2, 2, 3) # Lưới 2x2, vị trí 3
plt.imshow(new_img)
plt.title('Darkness Image')
```
