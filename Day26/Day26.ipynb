```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#c v hin th nh
img = cv2.imread('dataset/1.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img , cmap="gray")
plt.show()
```

```python
#c v hin th nh
img = cv2.imread('dataset/1.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Tng  sng
new_img = img.astype(np.float32)+50
new_img = np.clip(new_img , 0 , 255)
new_img = new_img.astype(np.uint8)

# Hin th hnh nh gc
plt.subplot(2, 2, 1) # Li 2x2, v tr 1
plt.imshow(img)
plt.title('Origin Image')

# Hin th hnh nh tng  sng
plt.subplot(2, 2, 2) # Li 2x2, v tr 2
plt.imshow(new_img)
plt.title('Brightness Image')

new_img = img.astype(np.float32)-80
new_img = np.clip(new_img , 0 , 255)
new_img = new_img.astype(np.uint8)

# Hin th hnh nh gim  sng
plt.subplot(2, 2, 3) # Li 2x2, v tr 3
plt.imshow(new_img)
plt.title('Darkness Image')
```
