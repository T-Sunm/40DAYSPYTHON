```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
filename= librosa.example('nutcracker') 
audio , sr = librosa.load(filename)

# audio.shape trả về 1 tupple và truy cập phần tử [0] lấy số lượng phần tử có trong mảng
print(audio.shape[0],sr)

"""
Bắt đầu (start): 0 giây (bắt đầu của file âm thanh).
Kết thúc (stop): Độ dài tổng cộng của file âm thanh tính bằng giây (được tính toán từ tổng số mẫu chia cho tần số lấy mẫu).
Số lượng điểm (num): Bằng với tổng số mẫu trong file âm thanh, cho phép mỗi mẫu có một điểm thời gian tương ứng.
"""

# điểm bắt đầu của trục y là 0 , điểm lớn nhất của trục y là audio.shape[0] / sr
""" audio.shape[0] là tổng số mẫu âm thanh trong mảng audio.
    sr là tần số lấy mẫu, tức là số lần mẫu âm thanh được ghi lại mỗi giây.
"""
time = np.linspace(0 , audio.shape[0] / sr  , num = audio.shape[0])
print(time)

plt.plot(time,audio,color="blue")
plt . xlabel ("Time (s)")
plt . ylabel ("Amplitude (quantized)")
plt . title ("Wav file visualization")
plt . axhline ( y =0 , color ='r', linestyle ='-')
plt . show ()
```

```python
from gtts import gTTS
import IPython
test_case_2 = "Tôi yêu AIO "
lang = 'vi'
output_filename = 'record.mp3'

output = gTTS ( test_case_2,lang = lang,slow = False) # text to speech

output.save(output_filename) # save google audio to a file
data,sr = librosa.load ( output_filename ) # load google audio using librosa library

IPython.display.display( IPython.display.Audio(np.transpose(data),
rate = sr ) ) # display audio
```
